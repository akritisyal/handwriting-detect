```python
import pandas as pd 
set= pd.read_csv("set.csv") 
```

```python
set.shape
```

```python
import glob
label=[]
for im in glob.glob('dataset/*/*'):
   l=(im.split('.')[0].split('\\')[1])
   label.append(l)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
import glob
label=[]
a=[]
for im in glob.glob('dataset/*/*'):
   l=(im.split('.')[0].split('\\')[1])
   label.append(l)
    
r=[0.5,0.4,0.3,0.2,0.1]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    gnb = GaussianNB()
    gnb.fit(X_train, y_train)

    y_pred = gnb.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    a.append(metrics.accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np
n_groups=5
index = np.arange(n_groups)

plt.plot(a)
plt.xticks(index+0.15, ('50:50', '60:40', '70:30', '80:20', '90:10'))
plt.title('Accuracy for Different size test data for Naive Bayes')
plt.ylabel('Accuracy')
plt.xlabel('Configuration')
```

```python
from sklearn.neighbors import KNeighborsClassifier 
r=[0.5,0.4,0.3,0.2,0.1]
b=[]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(metrics.accuracy_score(y_test,y_pred))
    b.append(metrics.accuracy_score(y_test, y_pred))
```

```python
n_groups=5
index = np.arange(n_groups)

plt.plot(b)
plt.xticks(index+0.15, ('50:50', '60:40', '70:30', '80:20', '90:10'))
plt.title('Accuracy for Different size test data for KNN')
plt.ylabel('Accuracy')
plt.xlabel('Configuration')
```

```python
from sklearn import svm
r=[0.5,0.4,0.3,0.2,0.1]
c=[]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    clf = svm.SVC(gamma='scale')
    clf.fit(X_train, y_train) 
    y_pred = clf.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    c.append(metrics.accuracy_score(y_test, y_pred))
```

```python
n_groups=5
index = np.arange(n_groups)

plt.plot(c)
plt.xticks(index+0.15, ('50:50', '60:40', '70:30', '80:20', '90:10'))
plt.title('Accuracy for Different size test data for SVM ')
plt.ylabel('Accuracy')
plt.xlabel('Configuration')
```

```python
from sklearn.ensemble import GradientBoostingClassifier
r=[0.5,0.4,0.3,0.2,0.1]
d=[]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0).fit(X_train, y_train)
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

    d.append(metrics.accuracy_score(y_test, y_pred))
```

```python
n_groups=5
index = np.arange(n_groups)

plt.plot(d)
plt.xticks(index+0.15, ('50:50', '60:40', '70:30', '80:20', '90:10'))
plt.title('Accuracy for Different size test data for Gradient Boosting')
plt.ylabel('Accuracy')
plt.xlabel('Configuration')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
r=[0.5,0.4,0.3,0.2,0.1]
e=[]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    clf = RandomForestClassifier(n_estimators=100, max_depth=3,random_state=0)
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

    e.append(metrics.accuracy_score(y_test, y_pred))
```

```python
n_groups=5
index = np.arange(n_groups)

plt.plot(e)
plt.xticks(index+0.15, ('50:50', '60:40', '70:30', '80:20', '90:10'))
plt.title('Accuracy for Different size test data for Random Forest ')
plt.ylabel('Accuracy')
plt.xlabel('Configuration')
```

```python
k_range=[1,3,5,7,9]
X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=0.4,random_state=4)
scores=[]
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(metrics.accuracy_score(y_test,y_pred))
    scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
plt.plot(k_range, scores)
plt.title('Accuracy for Different values of k for KNN')
plt.xlabel('Value of K for KNN')
plt.ylabel('Accuracy')
```
