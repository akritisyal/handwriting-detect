```python
import numpy as np
set=np.column_stack((a,b,c))
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier 
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd 
```

```python
import glob
label=[]
for im in glob.glob('dataset/*/*'):
    l=(im.split('.')[0].split('\\')[1])
    label.append(l)
```

```python
r=[0.5,0.4,0.3,0.2,0.1]
t=[]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(metrics.accuracy_score(y_test,y_pred))
    t.append(metrics.accuracy_score(y_test, y_pred))
```

```python
r=[0.5,0.4,0.3,0.2,0.1]
a=[]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=i,random_state=4)
    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(metrics.accuracy_score(y_test,y_pred))
    a.append(metrics.accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np
n_groups=5
index = np.arange(n_groups)

plt.plot(a,label='all attributes')
plt.plot(t,label='Blob')
plt.xticks(index+0.15, ('50:50', '60:40', '70:30', '80:20', '90:10'))
plt.title('Accuracy of Blob and All attribtes with different size test data for KNN')
plt.ylabel('Accuracy')
plt.xlabel('Configuration')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
y=[]
r=[0.5,0.4,0.3,0.2,0.1]
for i in r:
    X_train, X_test, y_train, y_test = train_test_split(set, label, test_size=i,random_state=4)
    gnb = GaussianNB()
    gnb.fit(X_train, y_train)

    y_pred = gnb.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    y.append(metrics.accuracy_score(y_test, y_pred))
```
